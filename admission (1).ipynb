```python
X=df.drop(['Chance of Admit '],axis=1)
X
```

```python
Y=df['Chance of Admit ']
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=42)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
X_train=pd.DataFrame(X_train,columns=X.columns)
X_test=pd.DataFrame(X_test,columns=X.columns)
```

```python
#modeling data(learning machine)
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor


#SVR model
svr_model = SVR(kernel='linear')  # You can specify different kernels like 'linear', 'poly', 'rbf', etc.
svr_model.fit(X_train, Y_train)



# DecisionTreeRegressor model
tree_model = DecisionTreeRegressor(max_depth=5)  # You can specify different parameters like 'max_depth', 'min_samples_split', etc.
tree_model.fit(X_train, Y_train)


# RandomForestRegressor model
rf_model = RandomForestRegressor(n_estimators=100, max_depth=5)  # You can specify different parameters like 'n_estimators', 'max_depth', etc.
rf_model.fit(X_train, Y_train)


#GradientBoostingRegressor model
gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=5)
gb_model.fit(X_train, Y_train)


# Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train,Y_train)


# Ridge Regression
ridge_model = Ridge(alpha=0.5)  # Alpha is the regularization strength
ridge_model.fit(X_train, Y_train)

# Lasso Regression
lasso_model = Lasso(alpha=0.1)  # Alpha is the regularization strength
lasso_model.fit(X_train,Y_train)
```

```python
#testing(machine test)
Y_svr=svr_model.predict(X_test)
Y_rf=rf_model.predict(X_test)
Y_tree=tree_model.predict(X_test)
Y_gb=gb_model.predict(X_test)
Y_linear=linear_model.predict(X_test)
Y_lasso=lasso_model.predict(X_test)
Y_ridge=ridge_model.predict(X_test)
```

```python
#accuracycheck(correction)
from sklearn.metrics import mean_squared_error
print("svr :",mean_squared_error(Y_test,Y_svr))
print("RF :",mean_squared_error(Y_test,Y_rf))
print("GB :",mean_squared_error(Y_test,Y_gb))
print("TREE :",mean_squared_error(Y_test,Y_tree))
print("LINEAR :",mean_squared_error(Y_test,Y_linear))
print("LASSO :",mean_squared_error(Y_test,Y_lasso))
print("RIDGE :",mean_squared_error(Y_test,Y_ridge))
```
